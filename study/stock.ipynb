```python
import os
import io
from dotenv import load_dotenv
import requests
import zipfile

# .env 파일 로드
load_dotenv()
api_key = os.getenv("opendart_key")
```

```python

# 예제 API 호출 (requests 사용)

response = requests.get(
    "https://opendart.fss.or.kr/api/corpCode.xml",
    params={"crtfc_key": api_key}
)
```

```python
response
```

```python
zip_file = zipfile.ZipFile(io.BytesIO(response.content))

# 4. ZIP 내부 파일 목록 출력
print("ZIP 파일 안에 들어 있는 파일들:")
print(zip_file.namelist())
```

```python
zip_file.namelist()[0]
```

```python
zip_file.extractall()
```
